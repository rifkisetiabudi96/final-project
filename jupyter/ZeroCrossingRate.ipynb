```python
import matplotlib.pyplot as plt
import numpy as np
import librosa
import librosa.display
import IPython.display as ipd
```

```python
# Loading audio files
sample1_file = "F:/FINALPROJECT/audio/01.wav"
sample2_file = "F:/FINALPROJECT/audio/02.wav"
sample3_file = "F:/FINALPROJECT/audio/03.wav"
```

```python
# Use Librosa To Read Audio
sample1, _ = librosa.load(sample1_file)
sample2, _ = librosa.load(sample2_file)
sample3, _ = librosa.load(sample3_file)
```

```python
# Determine Frame Size & Hop Length
FRAME_SIZE = 1024
HOP_LENGTH = 512
```

```python
# Create ZCR Variable
zcr_sample1 = librosa.feature.zero_crossing_rate(sample1, frame_length=FRAME_SIZE, hop_length=HOP_LENGTH)[0]
zcr_sample2 = librosa.feature.zero_crossing_rate(sample2, frame_length=FRAME_SIZE, hop_length=HOP_LENGTH)[0]
zcr_sample3 = librosa.feature.zero_crossing_rate(sample3, frame_length=FRAME_SIZE, hop_length=HOP_LENGTH)[0]
```

```python
# Check ZCR for Sample 1
zcr_sample1.size
```

```python
# Check ZCR for Sample 2
zcr_sample2.size
```

```python
# Check ZCR for Sample 3
zcr_sample3.size
```

```python
# Create t variable
frames = range(len(zcr_sample3))
t = librosa.frames_to_time(frames, hop_length=HOP_LENGTH)
```

```python
# Visualize ZCR With Librosa
plt.figure(figsize=(15, 10))

plt.plot(t, zcr_sample1, color="y")
#plt.plot(t, zcr_sample2, color="r")
#plt.plot(t, zcr_sample3, color="b")
plt.xlim(0, 14)
plt.ylim(0, 0.4)
plt.show()

```
