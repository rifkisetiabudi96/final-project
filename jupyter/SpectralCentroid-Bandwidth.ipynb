```python
import matplotlib.pyplot as plt
import numpy as np
import librosa
import IPython.display as ipd
```

```python
# Loading audio files
sample1_file = "F:/FINALPROJECT/audio/01.wav"
sample2_file = "F:/FINALPROJECT/audio/02.wav"
sample3_file = "F:/FINALPROJECT/audio/03.wav"
```

```python
# load audio files with librosa
sample1, sl = librosa.load(sample1_file)
sample2, _ = librosa.load(sample2_file)
sample3, _ = librosa.load(sample3_file)
```

```python
# Spectral centroid with Librosa
FRAME_SIZE = 1024
HOP_LENGTH = 512
```

```python
sc_sample1 = librosa.feature.spectral_centroid(y=sample1, sr=sl, n_fft=FRAME_SIZE, hop_length=HOP_LENGTH)[0]
sc_sample2 = librosa.feature.spectral_centroid(y=sample2, sr=sl, n_fft=FRAME_SIZE, hop_length=HOP_LENGTH)[0]
sc_sample3 = librosa.feature.spectral_centroid(y=sample3, sr=sl, n_fft=FRAME_SIZE, hop_length=HOP_LENGTH)[0]
```

```python
sc_sample1.shape
```

```python
# Visualising spectral centroid
frames = range(len(sc_sample1))
t = librosa.frames_to_time(frames, hop_length=HOP_LENGTH)
```

```python
len(t)
```

```python
# Show Spectral Centroid
plt.figure(figsize=(25,10))

plt.plot(t, sc_sample1, color='b')
#plt.plot(t, sc_sample2, color='r')
#plt.plot(t, sc_sample3, color='y')
plt.xlim(0, 14)

plt.show()
```

```python
# Spectral bandwidth with Librosa
sb_sample1 = librosa.feature.spectral_bandwidth(y=sample1, sr=sl, n_fft=FRAME_SIZE, hop_length=HOP_LENGTH)[0]
sb_sample2 = librosa.feature.spectral_bandwidth(y=sample2, sr=sl, n_fft=FRAME_SIZE, hop_length=HOP_LENGTH)[0]
sb_sample3 = librosa.feature.spectral_bandwidth(y=sample3, sr=sl, n_fft=FRAME_SIZE, hop_length=HOP_LENGTH)[0]
```

```python
sb_sample1.shape
```

```python
# Visualising spectral bandwidth
plt.figure(figsize=(25,10))

plt.plot(t, sb_sample1, color='b')
#plt.plot(t, sb_sample2, color='r')
#plt.plot(t, sb_sample3, color='y')
plt.xlim(0, 14)
plt.ylim(0, 3500)
plt.show()
```
