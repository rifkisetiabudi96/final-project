```python
import librosa
import librosa.display
import IPython.display as ipd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Loading audio files
sample1_file = "F:/FINALPROJECT/audio/01.wav"
```

```python
# load audio files with librosa
sample1, sl = librosa.load(sample1_file)
```

```python
# Extracting MFCCs
mfccs = librosa.feature.mfcc(y=sample1, n_mfcc=13, sr=sl)
mfccs.shape
```

```python
# Extracting MFCCs
plt.figure(figsize=(25, 10))
librosa.display.specshow(mfccs, 
                         x_axis="time", 
                         sr=sl)
plt.colorbar(format="%+2.f")
plt.show()
```

```python
# Computing first / second MFCCs derivatives
delta_mfccs = librosa.feature.delta(mfccs)
delta2_mfccs = librosa.feature.delta(mfccs, order=2)
delta_mfccs.shape
```

```python
# Show First Derivatives
plt.figure(figsize=(25, 10))
librosa.display.specshow(delta_mfccs, 
                         x_axis="time", 
                         sr=sl)
plt.colorbar(format="%+2.f")
plt.show()
```

```python
# Show Second Derivatives
plt.figure(figsize=(25, 10))
librosa.display.specshow(delta2_mfccs, 
                         x_axis="time", 
                         sr=sl)
plt.colorbar(format="%+2.f")
plt.show()
```

```python
mfccs_features = np.concatenate((mfccs, delta_mfccs, delta2_mfccs))
```
