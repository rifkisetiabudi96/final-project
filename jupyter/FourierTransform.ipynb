```python
import librosa
import librosa.display
import scipy as sp
import IPython.display as ipd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Loading audio files
sample1_file = "F:/FINALPROJECT/audio/01.wav"
```

```python
# Load Audio File
signal1, sr1 = librosa.load(sample1_file)
```

```python
# plot waveform
plt.figure(figsize=(18, 8))
librosa.display.waveplot(signal1, sr = sr1, alpha=0.5)
plt.show()
```

```python
# Derive spectrum using FT
ft = sp.fft.fft(signal1)
magnitude = np.absolute(ft)
frequency = np.linspace(0, sr1, len(magnitude))
```

```python
# plot spectrum
plt.figure(figsize=(18, 8))
plt.plot(frequency[:5000], magnitude[:5000]) # magnitude spectrum
plt.xlim(0, 400)
plt.ylim(0, 3000)
plt.xlabel("Frequency (Hz)")
plt.ylabel("Magnitude")
plt.show()
```
