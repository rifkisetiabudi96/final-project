```python
import os
import librosa
import librosa.display
import IPython.display as ipd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Loading audio files
sample1_file = "F:/FINALPROJECT/audio/01.wav"
sample2_file = "F:/FINALPROJECT/audio/02.wav"
sample3_file = "F:/FINALPROJECT/audio/03.wav"
```

```python
# Use Librosa To Read Audio
sample1, sm = librosa.load(sample1_file)
sample2, _ = librosa.load(sample2_file)
sample3, _ = librosa.load(sample3_file)
```

```python
# Define Frame & Hope Size
FRAME_SIZE = 2048
HOP_SIZE = 512
```

```python
# Create STFT for sample (Each sample will made same result)
S_sample1 = librosa.stft(sample1, n_fft=FRAME_SIZE, hop_length=HOP_SIZE)

S_sample1.shape
```

```python
# Analyze type of STFT
type(S_sample1[0][0])
```

```python
# Calculating Spectogram
Y_sample1 = np.abs(S_sample1) ** 2
```

```python
Y_sample1.shape
```

```python
type(Y_sample1[0][0])
```

```python
# Visualizing the spectrogram
def plot_spectrogram(Y, sr, hop_length, y_axis="linear"):
    plt.figure(figsize=(25, 10))
    librosa.display.specshow(Y, 
                             sr=sm, 
                             hop_length=hop_length, 
                             x_axis="time", 
                             y_axis=y_axis)
    plt.colorbar(format="%+2.f")
```

```python
plot_spectrogram(Y_sample1, sm, HOP_SIZE)
```

```python
# Log-Amplitude Spectrogram
Y_log_sample1 = librosa.power_to_db(Y_sample1)
plot_spectrogram(Y_log_sample1, sm, HOP_SIZE)
```

```python
# Visualising samples

S_sample1 = librosa.stft(sample1, n_fft=FRAME_SIZE, hop_length=HOP_SIZE)
S_sample2 = librosa.stft(sample2, n_fft=FRAME_SIZE, hop_length=HOP_SIZE)
S_sample3 = librosa.stft(sample3, n_fft=FRAME_SIZE, hop_length=HOP_SIZE)


Y_sample1 = librosa.power_to_db(np.abs(S_sample1) ** 2)
Y_sample2 = librosa.power_to_db(np.abs(S_sample2) ** 2)
Y_sample3 = librosa.power_to_db(np.abs(S_sample3) ** 2)

plot_spectrogram(Y_sample1, sm, HOP_SIZE, y_axis="log")
plot_spectrogram(Y_sample2, sm, HOP_SIZE, y_axis="log")
plot_spectrogram(Y_sample3, sm, HOP_SIZE, y_axis="log")
plt.show(block=False)
```
